```python
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df2010 =  pd.read_csv('csv/Censo_Escolas_MG_2010.csv', sep=','  , engine='python')
print('Contents of Dataframe : ')
df2010
```

```python
df2018 =  pd.read_csv('csv/Censo_Escolas_MG_2018.csv', sep=','  , engine='python')
df2018
```

```python
df2010_full = df2010[np.isfinite(df2010['TP_SITUACAO_FUNCIONAMENTO'])]
df2018_full = df2018[np.isfinite(df2018['TP_SITUACAO_FUNCIONAMENTO'])]

abertas2010 = df2010_full[df2010_full['TP_SITUACAO_FUNCIONAMENTO'] == 1] #17k
abertas2010
```

```python
fechada2018 = df2018['TP_SITUACAO_FUNCIONAMENTO'] == 3
fechada_ate2018 = df2018['TP_SITUACAO_FUNCIONAMENTO'] == 4
paralisada2018 = df2018['TP_SITUACAO_FUNCIONAMENTO'] == 2
naberta2018 = df2018[ fechada2018 | fechada_ate2018 | paralisada2018 ] #14k
naberta2018
```

```python
merged = naberta2018.merge(abertas2010[['CO_ENTIDADE']], on = 'CO_ENTIDADE')
merged
# save resulting output file  
```
