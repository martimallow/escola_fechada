```python
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df2010 =  pd.read_csv('csv/Censo_Escolas_MG_2010.csv', sep=','  , engine='python')
print('Contents of Dataframe : ')
```

```python
df2018 =  pd.read_csv('csv/Censo_Escolas_MG_2018.csv', sep=','  , engine='python')
df2018
```

```python
df2010_full = df2010[np.isfinite(df2010['TP_SITUACAO_FUNCIONAMENTO'])]
df2018_full = df2018[np.isfinite(df2018['TP_SITUACAO_FUNCIONAMENTO'])]

abertas2010 = df2010_full[df2010_full['TP_SITUACAO_FUNCIONAMENTO'] == 1] #17k
abertas2010
```

```python
abertas2018 = df2018_full[df2018_full['TP_SITUACAO_FUNCIONAMENTO'] == 1] 
abertas2018
```

```python
abertas2010['infra'] = abertas2010.IN_AGUA_FILTRADA+abertas2010.IN_ENERGIA_REDE_PUBLICA
+abertas2010.IN_ESGOTO_REDE_PUBLICA+abertas2010.IN_LIXO_COLETA_PERIODICA
+abertas2010.IN_LIXO_RECICLA+abertas2010.IN_SALA_DIRETORIA+abertas2010.IN_SALA_PROFESSOR
+abertas2010.IN_LABORATORIO_INFORMATICA+abertas2010.IN_LABORATORIO_CIENCIAS
+abertas2010.IN_SALA_ATENDIMENTO_ESPECIAL+abertas2010.IN_QUADRA_ESPORTES+abertas2010.IN_COZINHA+abertas2010.IN_BIBLIOTECA
+abertas2010.IN_SALA_LEITURA+abertas2010.IN_PARQUE_INFANTIL+abertas2010.IN_BERCARIO+abertas2010.IN_BANHEIRO_FORA_PREDIO
+abertas2010.IN_BANHEIRO_DENTRO_PREDIO+abertas2010.IN_BANHEIRO_EI+abertas2010.IN_BANHEIRO_PNE
+abertas2010.IN_DEPENDENCIAS_PNE

fig = plt.figure(figsize = (20, 10))
fig.patch.set_facecolor('xkcd:mint green')
title = fig.suptitle("Indice de Infraestrura", fontsize=25)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax = fig.add_subplot(1,1, 1)
ax.set_xlabel("Nivel de Infraestrutura",fontsize=20)
ax.set_ylabel("Quantidade de Escolass",fontsize=20) 
w_q = abertas2010['infra'].value_counts()
w_q = (list(w_q.index), list(w_q.values))
ax.tick_params(axis='both', which='major', labelsize=14)
bar = ax.bar(w_q[0], w_q[1], color='steelblue', 
        edgecolor='black', linewidth=2)
```
